```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = [1, 2, 3, 4, 5, 6] #x values
ys = [5, 4, 6, 5, 6, 7] #y values
plt.style.use("ggplot") #graph theme(style)
plt.plot(xs, ys) #first graph
plt.show()
plt.scatter(xs, ys) #second graph
plt.show()
```

```python
xs = np.array([1, 2, 3, 4, 5, 6])
ys = np.array([5, 4, 6, 5, 6, 7])

def slope(xs, ys):
    #slope formula
    m = (((np.mean(xs)*np.mean(ys)) - np.mean(xs*ys)) / 
         ((np.mean(xs)**2) - np.mean(xs**2)))
    #intercept formula
    b = np.mean(ys) - m * np.mean(xs)
    return m,b

m, b = slope(xs, ys)
#regression line (mx + b)
regression_line = [(m*x)+b for x in xs]

plt.scatter(xs, ys)
plt.plot(xs, regression_line)
plt.show()
```
